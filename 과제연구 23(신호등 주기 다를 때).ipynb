```python
def fast_traffic(a,b,c,d,e,f,g,h):
    traf=[traffic1(a,b,c,d,e,f,g,h),traffic2(a,b,c,d,e,f,g,h),traffic3(a,b,c,d,e,f,g,h),traffic4(a,b,c,d,e,f,g,h),traffic5(a,b,c,d,e,f,g,h),traffic6(a,b,c,d,e,f,g,h),traffic7(a,b,c,d,e,f,g,h),traffic8(a,b,c,d,e,f,g,h)]
    return min(traf)
def path_traffic(a,b,c,d,e,f,g,h):
    path=['e-a-b','g-h-d','e-f-d','e-c-d','g-h-f-a-b','g-h-c-a-b','e-f-c-a-b','g-h-f-c-d']
    traf=[traffic1(a,b,c,d,e,f,g,h),traffic2(a,b,c,d,e,f,g,h),traffic3(a,b,c,d,e,f,g,h),traffic4(a,b,c,d,e,f,g,h),traffic5(a,b,c,d,e,f,g,h),traffic6(a,b,c,d,e,f,g,h),traffic7(a,b,c,d,e,f,g,h),traffic8(a,b,c,d,e,f,g,h)]
    temp = fast_traffic(a,b,c,d,e,f,g,h)
    for i in range(8):
        if temp==traf[i]:
            yield path[i]
def print_path_traffic(a,b,c,d,e,f,g,h):
    for i in path_traffic(a,b,c,d,e,f,g,h):
        print(i)
def traffic_time(a,b,c,d,e,f,g,h):
    print(traffic1(a,b,c,d,e,f,g,h))
    print(traffic2(a,b,c,d,e,f,g,h))
    print(traffic3(a,b,c,d,e,f,g,h))
    print(traffic4(a,b,c,d,e,f,g,h))
    print(traffic5(a,b,c,d,e,f,g,h))
    print(traffic6(a,b,c,d,e,f,g,h))
    print(traffic7(a,b,c,d,e,f,g,h))
    print(traffic8(a,b,c,d,e,f,g,h))
```
