```python
def trifast_traffic(a,b,c,d,e,f,g,h):
    traf=[tritraffic1(a,b,c,d,e,f,g,h),tritraffic2(a,b,c,d,e,f,g,h),tritraffic3(a,b,c,d,e,f,g,h),tritraffic4(a,b,c,d,e,f,g,h)]
    return min(traf)

def tripath_traffic(a,b,c,d,e,f,g,h):
    num=['e-a-b','g-h-d','e-f-d','e-c-d']
    traf=[tritraffic1(a,b,c,d,e,f,g,h),tritraffic2(a,b,c,d,e,f,g,h),tritraffic3(a,b,c,d,e,f,g,h),tritraffic4(a,b,c,d,e,f,g,h)]
    temp = trifast_traffic(a,b,c,d,e,f,g,h)
    for i in range(4):
        if temp==traf[i]:
            yield num[i]
```
